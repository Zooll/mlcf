```python
%matplotlib notebook 
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_TRAIN_PATH = 'blood-pressure-estimation-data/data_train'
DATA_TRAIN_ANS_PATH = 'blood-pressure-estimation-data/data_train.ans'
```

```python
def load_answers(path):
    sbp_dbp = {}
    with open(path, 'r') as f:
        for line in f:
            vals = line[:-1].split(',')
            sbp_dbp[vals[0]] = (float(vals[1]), float(vals[2]))
    return sbp_dbp

sbp_dbp = load_answers(DATA_TRAIN_ANS_PATH)

def read_ppg_ecg(filename):
    ppg = []
    ecg = []
    with open(DATA_TRAIN_PATH + '/' + filename, 'r') as f:
        for line in f:
            vals = line[:-1].split(',')
            ppg.append(vals[0])
            ecg.append(vals[1])
    return ppg, ecg
```

```python
def draw_graphics(names = None, count = 1):
    if names is None:
        names = os.listdir(DATA_TRAIN_PATH)
    names = names[:count]
    
    for name in names:
        ppg, ecg = read_ppg_ecg(name)
        fig, ax = plt.subplots()
        ax.set_title(name + ' PPG')
        ax.plot(ppg[1:])
        fig2, ax2 = plt.subplots() 
        ax2.set_title(name + ' ECG')
        ax2.plot(ecg[1:])
        plt.show()
```

```python
draw_graphics(count = 1)
```
